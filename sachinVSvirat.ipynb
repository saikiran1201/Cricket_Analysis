```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import pandas as pd
tables_sachin = pd.read_html('https://www.espncricinfo.com/india/content/player/35320.html')
tables_virat = pd.read_html('https://www.espncricinfo.com/india/content/player/253802.html')
```

```python
len(tables_sachin) #tables
```

```python
len(tables_virat) #tables
```

```python
df_sachin = tables_sachin[0] 
df_sachin
```

```python
df_sachin = df_sachin.rename(columns={'Unnamed: 0':'Type'},inplace=None)
df_sachin
```

```python
df_virat = tables_virat[0] 
df_virat
```

```python
df_virat = df_virat.rename(columns={'Unnamed: 0':'Type'},inplace=None)
df_virat
```

```python
fig = plt.figure()
X = np.arange(6)
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, df_virat['Runs'],color = 'r', width = 0.25)
ax.bar(X + 0.25, df_sachin['Runs'],color = 'b', width = 0.25)

#legend
plt.title("Type of match VS Runs scored")
plt.xlabel ('Type of match')
plt.ylabel ('Runs')
plt.legend(["Virat Kohli", "Sachin Tendulkar"], loc ="upper right") 

x = np.array(df_virat['Type'])
ax.set_xticks(X)
ax.set_xticklabels(x)


for p in ax.patches:

    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
fig = plt.figure()
X = np.arange(6)
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0, df_virat['Ave'], color = 'r', width = 0.25)
ax.bar(X + 0.25, df_sachin['Ave'], color = 'b', width = 0.25)

plt.title("Type of match VS Average")
plt.xlabel ('Type of match')
plt.ylabel ('Average')
plt.legend(["Virat Kohli", "Sachin Tendulkar"], loc ="upper right") 

x = np.array(df_virat['Type'])
ax.set_xticks(X)
ax.set_xticklabels(x)

for p in ax.patches:

    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
fig = plt.figure()
X = np.arange(6)
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, df_virat['100'], color = 'r', width = 0.25)
ax.bar(X + 0.25, df_sachin['100'], color = 'b', width = 0.25)

plt.title("Type of match VS Number of centuries")
plt.xlabel ('Type of match')
plt.ylabel ('Number of centuries')
plt.legend(["Virat Kohli", "Sachin Tendulkar"], loc ="upper right") 


x = np.array(df_virat['Type'])
ax.set_xticks(X)
ax.set_xticklabels(x)

for p in ax.patches:

    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
