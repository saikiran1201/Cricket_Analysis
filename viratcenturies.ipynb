```python
import pandas as pd
import matplotlib.pyplot as plt
tables_virat100 = pd.read_html('https://www.sportsadda.com/cricket/features/68-international-centuries-11-years-1-virat-kohli')
```

```python
tables_virat100[0]
```

```python
df_virat100 = tables_virat100[0] 
df_virat100
```

```python
df_virat100.columns = ['Years', '100s', 'innings per 100']
```

```python
df_virat100.loc[0]
```

```python
df_virat100 = df_virat100.drop([0])
```

```python
df_virat100 = df_virat100.drop('innings per 100',axis=1)
```

```python
df_virat100
```

```python
#df_virat100.plot.line()
df_virat100 = df_virat100.astype(int)
ax = df_virat100.plot.line(x ='Years', y = '100s')
plt.xlabel ('YEARS')
plt.ylabel ('Number of centuries')
```

```python
ax = df_virat100.plot.bar(x ='Years', y = '100s')

for p in ax.patches:

    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))

plt.xlabel ('YEARS')
plt.ylabel ('Number of centuries')
```
