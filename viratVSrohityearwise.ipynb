```python
import pandas as pd
import matplotlib.pyplot as plt
tables = pd.read_html('https://scroll.in/field/958439/the-arcs-of-virat-kohli-and-rohit-sharmas-wildly-contrasting-odi-careers-are-finally-meeting')
```

```python
df_rohit = tables[1] 
df_rohit
```

```python
df_virat = tables[2] 
df_virat
```

```python
#df_virat100.plot.line()
#df_virat = df_virat.astype(int)
fig, ax = plt.subplots() 

ax.plot(df_virat['Year'], df_virat['100s'], color = 'r')

ax.plot(df_rohit['Year'], df_rohit['100s'], color = 'b')

ax.set_xlabel('YEARS')  
ax.set_ylabel('Number of centuries')  
plt.legend(["Virat Kohli", "Rohit Sharma"], loc ="upper right") 
ax.set_title("comparision of number of centuries")  
```

```python
fig, ax = plt.subplots() 

ax.plot(df_virat['Year'], df_virat['Runs'], color = 'r')

ax.plot(df_rohit['Year'], df_rohit['Runs'], color = 'b')

ax.set_xlabel('YEARS')  
ax.set_ylabel('Number of centuries')  
plt.legend(["Virat Kohli", "Rohit Sharma"], loc ="upper right") 
ax.set_title("comparision of number of centuries")  
```
