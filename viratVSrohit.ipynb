```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
import pandas as pd
tables_virat = pd.read_html('https://www.espncricinfo.com/india/content/player/253802.html')
tables_rohit = pd.read_html('https://www.espncricinfo.com/india/content/player/34102.html')
```

```python
fig = plt.figure()
X = np.arange(6)
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0, df_virat['Runs'], color = 'r', width = 0.25)
ax.bar(X + 0.25, df_rohit['Runs'], color = 'b', width = 0.25)

plt.title("Type of match VS Runs scored")
plt.xlabel ('Type of match')
plt.ylabel ('Runs')
plt.legend(["Virat Kohli", "Rohit Sharma"], loc ="upper right") 


x = np.array(df_virat['Type'])
ax.set_xticks(X)
ax.set_xticklabels(x)

for p in ax.patches:

    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
fig = plt.figure()
X = np.arange(6)
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0, df_virat['Ave'], color = 'r', width = 0.25)
ax.bar(X + 0.25, df_rohit['Ave'], color = 'b', width = 0.25)

plt.title("Type of match VS Average")
plt.xlabel ('Type of match')
plt.ylabel ('Average')
plt.legend(["Virat Kohli", "Rohit Sharma"], loc ="upper right") 

x = np.array(df_virat['Type'])
ax.set_xticks(X)
ax.set_xticklabels(x)

for p in ax.patches:

    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
